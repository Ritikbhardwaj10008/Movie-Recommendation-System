```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj): # object ke ander loop chalana hai kuki vo ek list hoga.
   # loop chlega to mujhe ek dictionary milega{},dictionary ke ander se i['name'] nikalna hai
   L.append(i['name'])


  return L
```

```python
# we see here ki isme name ke ander actor ka REAL name hai , so we have to take that name.
# hm upper wale function ko hi repeat krengae bs thoda se change krengae
def convert3(obj):
  L=[]
  counter =0
  for i in ast.literal_eval(obj): # object ke ander loop chalana hai kuki vo ek list hoga.
   # loop chlega to mujhe ek dictionary milega{},dictionary ke ander se i['name'] nikalna hai
   if counter !=3 : # agar counter 3 nhi hai to hm append karte jaengae
     L.append(i['name'])
     counter+=1
   else :
      break



  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director' :
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# for stemming we use nltk (natural language processing library)
import nltk
```

```python
from nltk.stem.porter import PorterStemmer  # for solution in action ,actions similar words.
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    ps.stem(i)
  return " ".join(y)  # convert to string  again
```

```python
# in dono ke beech ka similarity score calculate karna hai.
# yha pr need hai vectorisation ki .
# har movie ke tags ko ek vector mai convert krunga to har movie ek vector ban jaega
# to easy hoga similar movies dekhna .(closest vecotrs  )
# called text vectorisation -> technieques are Bag Of words,tfidf,wordtovec.
# here we use bag of words.
```

```python
# 1)in this pehle saare tags  ko combine kr dete hai (concatenate) (1 tag mai 100 words hai to sabko combine karne pr 50000 words ho jaye)
# tag1+tag2+tag3+   tagn
# 2) now usme se 5000 most common words chahiyae.(top 5000 words jinka frequency sabse zaiyada hai unko extract krunga)
# 3) now har movie ke tag ko fir se uthaunga ,1st movie mai puchunga ki ek word kitni bar aaya hai ,2nd word kitni bar aaya ,word 3 kitni bar aaya ,
# 4) 2nd movie ko pkdunga then again check 1-5000 words kitni bar ayae hai. (kitne bhi words le sakte hai 5000,6000,7000,10000),best ki kam words mai kam ho jaye.
# 5) end mai mere pass data frame aa ajega jiska shape hoga 5000,5000
# 6) vecotrisation mai hm 'stopwords' ka use nhi krengae.
# 7) in sklearn we have class CountVectoriser

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')  # maxfeatures means kitne words hme lene hai.,stop_word english language ke words ko hata dena hai
```

```python
# now hme har movie ka har dusri movie ke saath distance calculate karna hai .# distance jitna zaiyada hoga utna hi similarity km hoga.
# yha pr euclidean distance calculate nhi krengae hm cosine distance check krengae (means angle calculate krengae (circular))
# high dimension data jitna zaiyada high dimension data euclidean utna hi fail karta hai .
# so we use cosine distance .
# iske lie bhi ek function hai hmarre pass (cosine similarity)
from sklearn.metrics.pairwise import cosine_similarity # cosine_similarity bhi hai or cosine_distance bhi hai .(inverse of each other)
```

```python
similarity=cosine_similarity(vectors) # hm saare vectors ko pass krdengae (so har kisi ka ek dusre se distance nikl jaega
```

```python
# now hme ek function banana hai recommend bolke jisme agar koi mujhe ek movie deta hai to ye mujhe palat karke 5 similar moveis return kare.
#1:23
```

```python
# Now ab agar mujhe movie di jaye to mujhe index nikalna pdega then usse mai similarity matrix mai ghusunga
# or vo particular index ki movie nikalunga,then distances ko sort karunga->zaiyada similar movies pehle aa jayengae or km similar movies peeche chli jayengae
# then top 5 similar movies ko fetch karke dikha dunga
```

```python
# lets fetch the index
# new_df[new_df['title']=='Batman Begins'].index[0]  # index fetch karne ka taika .

def recommended(movie):
  movie_index=new_df[new_df['title']==movie].index[0]   # index 0 represents the column(in which index is written)
  distances=similarity[movie_index]   # distances nikalne hai (array hoga )
  # sort karna hai
  new_df[new_df['title']=='Batman Begins'].index[0]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]  # key is also a keyword of sorted.
  for i in movies_list:   # now movies ke list ko print kar dengae .
    print(new_df.iloc[i[0]].title)

  # sorting karne se index position loose hoga , iske lie hm hm enumerate function ko call krengae,
```
